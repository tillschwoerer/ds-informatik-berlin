# Wortvektoren

```python
import gensim.downloader
```

Wir laden Wortvektoren herunter, die mithilfe des Glove Models auf Wikipedia Daten trainiert wurden. Zu beachten ist, dass dieses Modell nur auf Englischen Daten trainiert wurde. Daher werden für die meisten deutschen Wörter keine Embeddings verfügbar sein.

```python
wv = gensim.downloader.load('glove-wiki-gigaword-100')
```

Der Wortvektor des Wortes Berlin ist ein 100-dimensionaler Vektor, der die Bedeutung von Berlin in einem kontinuierlichen Vektorraum repräsentiert.

```python
wv['berlin']
```

```python
print(len(wv['berlin']))
```

Was sind die ähnlichsten Wörter zu Berlin?

```python
wv.most_similar('berlin', topn=10)
```

Wir rechnen König - Mann + Frau und erhalten die ähnlichsten Wörter zu diesem Ergebnis.

```python
wv.most_similar(positive=['king', 'woman'], negative='man', topn=20)
```
